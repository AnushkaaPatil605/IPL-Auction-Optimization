```python
batsmen_score_df = batsmen_df.dropna(
    subset=['RunsScored','BattingAVG','BattingS/R']
).copy()
```

```python
batsmen_score_df.shape
```

```python
batsmen_score_df['Runs_Norm'] = (
    batsmen_score_df['RunsScored'] /
    batsmen_score_df['RunsScored'].max()
)

batsmen_score_df['Avg_Norm'] = (
    batsmen_score_df['BattingAVG'] /
    batsmen_score_df['BattingAVG'].max()
)

batsmen_score_df['SR_Norm'] = (
    batsmen_score_df['BattingS/R'] /
    batsmen_score_df['BattingS/R'].max()
)
```

```python
batsmen_score_df['PerformanceScore'] = (
    0.50 * batsmen_score_df['Runs_Norm'] +
    0.25 * batsmen_score_df['Avg_Norm'] +
    0.25 * batsmen_score_df['SR_Norm']
) * 100
```

```python
qualified_batsmen = batsmen_score_df[
    batsmen_score_df['RunsScored'] >= 1000
].copy()
```

```python
qualified_batsmen['Score_Category'] = pd.cut(
    qualified_batsmen['PerformanceScore'],
    bins=[0,50,75,100],
    labels=['Average','Good','Elite']
)
```

```python
qualified_batsmen[['Name','PerformanceScore','Score_Category']].head()
```

```python
qualified_batsmen[['Name','PerformanceScore','Score_Category']].head()
```

```python
qualified_batsmen['Price_Category'] = pd.cut(
    qualified_batsmen['ValueinCR'],
    bins=[0,5,10,20],
    labels=['Cheap','Moderate','Expensive']
)
```

```python
qualified_batsmen[['Name','ValueinCR','Price_Category']].head()
```

```python
qualified_batsmen[['Name',
                   'PerformanceScore',
                   'Score_Category',
                   'ValueinCR',
                   'Price_Category']].head(15)
```

```python
def classify_player(row):

    if row['Score_Category'] == 'Elite' and row['Price_Category'] == 'Cheap':
        return 'Hidden Gem'

    elif row['Score_Category'] == 'Elite' and row['Price_Category'] == 'Moderate':
        return 'Strong Buy'

    elif row['Score_Category'] == 'Elite' and row['Price_Category'] == 'Expensive':
        return 'Star Player'

    elif row['Score_Category'] == 'Average' and row['Price_Category'] == 'Expensive':
        return 'Overpriced'

    elif row['Score_Category'] == 'Good' and row['Price_Category'] == 'Cheap':
        return 'Value Pick'

    else:
        return 'Neutral'

qualified_batsmen['Recommendation'] = qualified_batsmen.apply(
    classify_player,
    axis=1
)
```

```python
qualified_batsmen[
[
'Name',
'PerformanceScore',
'ValueinCR',
'Recommendation'
]
].sort_values(
by='PerformanceScore',
ascending=False
)
```

```python
qualified_batsmen['MedianScore'] = qualified_batsmen['PerformanceScore'].median()

qualified_batsmen['MedianPrice'] = qualified_batsmen['ValueinCR'].median()
```

```python
qualified_batsmen[['MedianScore','MedianPrice']].head(1)
```

```python
def auction_category(row):

    if row['PerformanceScore'] >= 56.57 and row['ValueinCR'] < 7.5:
        return 'Hidden Gem'

    elif row['PerformanceScore'] >= 56.57 and row['ValueinCR'] >= 7.5:
        return 'Star Player'

    elif row['PerformanceScore'] < 56.57 and row['ValueinCR'] < 7.5:
        return 'Budget Pick'

    else:
        return 'Overpriced'


qualified_batsmen['AuctionCategory'] = qualified_batsmen.apply(
    auction_category,
    axis=1
)
```
